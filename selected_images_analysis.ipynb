## Understand the diagnosis distribution of selected images

The selected images are stored in `selected_img_dir`.

```python
import os
import pandas as pd
```

```python
selected_img_dir = '/local-scratch2/Datasets/Fitzpatrick17k/Selected/'
```

```python
all_df = pd.read_csv('fitzpatrick17k.csv', index_col=0, header='infer')
all_df.head()
```

```python
# https://stackoverflow.com/a/17071908
selected_img_ids = [f.split('.')[0] for f in os.listdir(selected_img_dir)]
print(len(selected_img_ids))

selected_df = all_df.loc[all_df['md5hash'].isin(selected_img_ids)]
```

```python
selected_df.groupby('label').size().reset_index()
```
